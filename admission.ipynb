```python
import numpy as np
import pandas as pd
```

```python
# Read training data
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
df_train.describe()
```

```python
df_train.groupby(['heart_disease']).agg(['count'])
```

```python
#One hot encoding
df_train_ohe = pd.get_dummies(df_train)
```

```python
# Imputing missing values
from sklearn.preprocessing import Imputer

imr = Imputer(missing_values=np.nan, strategy='mean')

patient_train_data = pd.DataFrame(imr.fit_transform(df_train_ohe))
patient_train_data.columns = df_train_ohe.columns
patient_train_data.index = df_train_ohe.index
patient_train_data.describe()
```

```python
# Choose target and features 
y = patient_train_data.admission_status
X = patient_train_data.drop(['id','admission_status'], axis=1)
X.head()
```

```python
ratio = float(np.sum(y == 0)) / np.sum(y == 1)
print(ratio)
```

```python
from sklearn.model_selection import train_test_split

# Holdout 20% data for model validation
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, stratify=y, random_state = 7)
print(float(np.sum(val_y == 0)) / np.sum(val_y == 1))
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

xgb = XGBClassifier(n_jobs=-1, scale_pos_weight = ratio)
param_grid = {
    "n_estimators": [50, 100, 500, 1000],
    "learning_rate": [0.05, 0.1, 0.5, 0.75],
#    "max_depth": [3,4,5],
#    "min_child_weight": [1,2],
#    "gamma": [0,1],
#    "subsample": [0.7,1],
#    "colsample_bytree": [0.7,1]
}

fit_params = {"eval_set": [(val_X, val_y)], 
              "early_stopping_rounds": 10, 
              "verbose": False} 

searchCV = GridSearchCV(xgb, cv=10, param_grid=param_grid, fit_params=fit_params)
searchCV.fit(train_X, train_y)
```

```python
searchCV.best_params_
```

```python
#Train the model with full training data
clf = searchCV.best_estimator_
clf.fit(X, y)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X, y, scoring='roc_auc', cv=10)
print(scores)
print(scores.mean())
```

```python
df_confusion = pd.crosstab(y, clf.predict(X), rownames=['Actual'], colnames=['Predicted'], margins=True)
```

```python
print(df_confusion)
```
